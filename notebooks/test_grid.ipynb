```python
from mcfs.fs_grid import run_gridded_tau
import os

base_root = "/home/STORAGE"
simulation_name = "TNG50-4"
num  = 25

nspec = 2
nbins = 1024
axis  = -1

lines = [
    ("H",  1, 1215),
    ("Si", 3, 1206),
    ("Si", 2, 1190),
    ("Si", 2, 1193),
]

savedir_base = "/home/STORAGE/SKEWERS/"
out_dir = os.path.join(savedir_base, simulation_name, f"snapdir_{num:03d}")

gs, info = run_gridded_tau(
    base=os.path.join(base_root, simulation_name),
    num=num,
    nspec=nspec,
    axis=axis,
    nbins=nbins,
    lines=lines,
    out_dir=out_dir,
    overwrite=True,
    MPI=None,
    kernel="voronoi",
    quiet=False,
)
print(info)
```
